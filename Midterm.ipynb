```python
# MIDTERM - Predict Class
import numpy as np # linear algebra
import matplotlib.pyplot as plt # For plotting
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
dataset = pd.read_csv(r'C:\Users\saedw\OneDrive\Desktop\CECS 456 Machine Learning\Data\Data_Midterm.csv')
# print snapshot of dataset
dataset
```

```python
# x needs to be a matrix for model - contain all predictor variables
x = dataset.iloc[:, [1, 2, 3, 4, 5, 6, 7, 8, 9]].values
y = dataset.iloc[:, 10].values
```

```python
# Splitting the dataset into the Training set and Test set
# TEST AND TRAINING MODELS USED FOR FOLLOWING MODELS
    # Kernels, Decision Trees, Random Forest, KNN, Logistic Regression, mabye XG Boosting
# Train = .80 Test = .20
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.20, random_state = 0)

# Feature Scaling - scale the data after training 
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# Gaussion Kernal Model - predicting Class (2 or 4)
# Fitting classifier to the Training set
from sklearn.ensemble import RandomForestClassifier
from sklearn.gaussian_process import GaussianProcessClassifier
from sklearn.gaussian_process.kernels import RBF
# define our kernel
kernel = 1.0 * RBF(1.0)
# selecting gaussian process classifier and fit model on training data
classifier_kernel = GaussianProcessClassifier(kernel=kernel, random_state=0).fit(X_train, y_train)

# Predicting the Test set results
# y_pred holds all of the predicted values for the data set
y_pred_kernel = classifier_kernel.predict(X_test)

# Confusion Matrix
print ("Confusion Matrix - Gaussian Kernal\n {}".format(confusion_matrix(y_test, y_pred_kernel)))
# Accuracy Score
print ("Accuracy score on Test data - Gaussian Kernal\n {}".format(accuracy_score(y_test, y_pred_kernel)))
```

```python
# RANDOM FOREST MODEL
# Fitting classifier to the Training set
from sklearn.ensemble import RandomForestClassifier
# selecting n_estimators - default is 10
classifierRandomForest = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 0)
# fit the model on training data sets X_train and y_train
classifierRandomForest.fit(X_train, y_train)

# Predicting the Test set results
# y_pred holds all of the predicted values for the data set
y_predRandomForest = classifierRandomForest.predict(X_test)


# Confusion Matrix
print ("Confusion Matrix - Random Forest Model\n {}".format(confusion_matrix(y_test, y_predRandomForest)))
# Accuracy Score
print ("Accuracy score on Test data - Random Forest Model\n {}".format(accuracy_score(y_test, y_predRandomForest)))
```

```python
# KNN MODEL
# Fitting classifier to the Training set
from sklearn.neighbors import KNeighborsClassifier
# k=5 neighbors
classifier_knn = KNeighborsClassifier(n_neighbors=5, metric='minkowski', p=2)
classifier_knn.fit(X_train, y_train)

# Predicting the Test set results
y_pred_knn = classifier_knn.predict(X_test)


# Confusion Matrix
print ("Confusion Matrix - KNN\n {}".format(confusion_matrix(y_test, y_pred_knn)))
# Accuracy Score
print ("Accuracy score on Test data  - KNN\n {}".format(accuracy_score(y_test, y_pred_knn)))
```

```python
# LOGISTIC REGRESSION MODEL
# Fit classifier to training set
from sklearn.linear_model import LogisticRegression
classifier_logistic = LogisticRegression()
classifier_logistic.fit(X_train, y_train)

# Predict Test results
y_pred_logistic = classifier_logistic.predict(X_test)

# Confusion Matrix
print("Confusion Matrix - Logistic Classification\n {}".format(confusion_matrix(y_test, y_pred_logistic)))
# Accuracy Score
print("Accuracy Score - Logistic Classification\n {}".format(accuracy_score(y_test, y_pred_logistic)))
```

```python
# Support Vector - K-fold cross validation and Grid Search

# Step 1 K-fold cross validation
from sklearn.model_selection import KFold 
from sklearn.svm import SVC



# x needs to be a matrix for model - contain all predictor variables
X = dataset.iloc[:, [1, 2, 3, 4, 5, 6, 7, 8, 9]].values
y = dataset.iloc[:, 10].values

k = 10
kf = KFold(n_splits=k, random_state=None)
model = SVC(kernel='linear', random_state=0)
 
acc_score = []
cm_holder = []
 
for train_index , test_index in kf.split(X):
    X_train , X_test = X[train_index,:],X[test_index,:]
    y_train , y_test = y[train_index] , y[test_index]
     
    model.fit(X_train,y_train)
    pred_values = model.predict(X_test)
     
    acc = accuracy_score(pred_values , y_test)
    acc_score.append(acc)
    
    cm = confusion_matrix(pred_values, y_test)
    cm_holder.append(cm)
     
avg_acc_score = sum(acc_score)/k
avg_confusion_matrix = sum(cm_holder)/k
 
print("Confusion Matrix - Support Vector (k-fold cross validation)\n".format(avg_confusion_matrix))
print('Avg accuracy : {}'.format(avg_acc_score))
```

```python
from xgboost import XGBClassifier

x = dataset.iloc[:, [1, 2, 3, 4, 5, 6, 7, 8, 9]].values
y = dataset.iloc[:, 10].values

y = (y-2)/2

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.20, random_state = 0)


# fit model no training data
classifier_xgb = XGBClassifier()
classifier_xgb.fit(X_train, y_train)


# make predictions for test data
y_predxgb = classifier_xgb.predict(X_test)

# Confusion Matrix
print("Confusion Matrix - XG Boost Classification\n {}".format(confusion_matrix(y_test, y_predxgb)))
# Accuracy Score
print("Accuracy Score - XG Boost Classification\n {}".format(accuracy_score(y_test, y_predxgb)))
```
